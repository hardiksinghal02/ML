```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.set_printoptions(precision=2)
```

```python
dataset = pd.read_csv("Position_Salaries.csv")

# No need to take 0th column as it is alerady encoded in column 1
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```


## Linear regression to compare

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()

linear_regressor.fit(X, y)
linear_regressor.predict([[3]])
```

## Polynomial regression model

```python
from sklearn.preprocessing import PolynomialFeatures

polynomial_regressor = PolynomialFeatures(degree=4)
X_poly = polynomial_regressor.fit_transform(X)

linear_regressor_2 = LinearRegression()
linear_regressor_2.fit(X_poly, y)
```

### Visualization

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, linear_regressor.predict(X), color = 'blue')

plt.title("Truth or bluff (LinearRegression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")

plt.show()
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, linear_regressor_2.predict(X_poly), color = 'blue')

plt.title("Truth or bluff (PolynomialRegression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")

plt.show()
```

```python
# Making multiple predictions
linear_regressor.predict([[6.5], [10]])
```
